```python
'''Assignment 1: Temperature Check
Problem:
Write a program that takes the current temperature as input and prints:
● "It's too cold!" if below 10°C
● "It's cool outside" if between 10°C and 25°C
● "It's hot outside!" if above 25°C'''




Current_Tempetature = int(input('enter the current temperature:'))

if Current_Tempetature<10:
    print("It's too cold!")
elif Current_Tempetature >= 10 and Current_Tempetature <= 25:
    print("It's cool outside")

else:
    print("It's hot outside!")


'''Assignment 2: ATM Withdrawal System
Problem:
Ask the user for their account balance and the amount they want to withdraw.
● If the withdrawal amount is greater than the balance → print 
● If the withdrawal amount is less than or equal to balance → print "Transaction
Successful" and show the remaining balance.'''
Balance = int(input('Enter the account balance:'))
Withdrawal = int(input('Enter the amount withdrawal'))


if Withdrawal>Balance:
    print("Insufficient Balance")


else:
    Balance>=Withdrawal

    print('Transaction Successfull')
    print('show the remaining balance')


'''
Assignment 3: Student Grade Calculator
Problem:
Input marks (0-100) and print grade based on:
● 90-100 → A+
● 80-89 → A
● 70-79 → B
● 60-69 → C
● Below 60 → Fail
'''

Marks = int(input('enter the student marks:'))

if Marks>=90 and Marks<=100:
    print('A+')
elif Marks>=80 and Marks<=89:
    print('A')
elif Marks>=70 and Marks<=79:
    print('B')
elif Marks>=60 and Marks<=69:
    print('c')
else:
    print('Fail')







'''Assignment 4: Bus Ticket Discount
Problem:
Ask the user for age and print ticket price:
● Below 5 → Free
● 5-18 → 50% Discount
● 60 or above → 30% Discount
● Others → Full Price'''


age = int(input('enter the user age:' ))



if age<5:
    print('Free')
elif age>=5 and age<=18:
    print('50% Discount')
elif age>=60:
    print('30% Discount')
else:
    print('Full Price')



'''Assignment 5: Restaurant Bill with Discount
Problem:
Take total bill amount from the user:
● If the bill is above 1000 → apply 10% discount
● Otherwise → no discount
Then print the final bill amount.'''


Bill = int(input('enter the bill amount:'))



if Bill>=1000:
    print("10% discount")
else:
    print("final bill amount") 




'''Problem:
Take a username and password input.
● If username == "admin" and password == "12345" → print "Login Successful"
● Else → print "Invalid Credentials"'''

username = input("enter the user name:" )
password = input("enter the password:" )

actual_username = "admin"
actual_password = "12345"

if username == actual_username and password == actual_password:
    print("Login Successful")
else:
    print("Invalid Credentials")

'''Assignment 7: Traffic Light System
Problem:
Ask the user to input a color (red, yellow, or green).
Print:
● "Stop" for red
● "Ready to go" for yellow
● "Go" for green
● "Invalid color" otherwise'''


color = input('enter the first color: ')



if color == 'red':
    print('stop')
elif color == 'yellow':
    print('Ready to go')
elif color == 'green':
    print('Go')
else:
    print('Invalid color')


'''Assignment 8: Leap Year Checker
Problem:
Take a year as input and check:
● If it is divisible by 400 → Leap Year
● Else if divisible by 4 but not by 100 → Leap Year
● Otherwise → Not a Leap Year'''



year = int(input('enter the year:'))

if year % 400 == 0:
    print('Leap Year')
elif year % 4 == 0 and year % 100 != 0:
    print('Leap year')
else:
    print('Not a Leap Year')

'''Assignment 10: Electricity Bill Calculator
Problem:
Input number of units consumed and calculate bill:
● Up to 100 units → 5 BDT per unit
● 101-200 units → 7 BDT per unit
● Above 200 units → 10 BDT per unit
Print total payable amount.'''

Bill = int(input('enter the units consumed:' ))

if Bill<=100:
    print('5 BDT per unit')
elif Bill >=101 and Bill<=200:
    print('7 BDT per unit')
else:
    print('10 BDT per unit')



```
